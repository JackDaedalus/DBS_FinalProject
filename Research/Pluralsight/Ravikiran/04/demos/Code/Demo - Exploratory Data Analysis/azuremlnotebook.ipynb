```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/home/nbuser/AdultDataset.csv')
```

```python
data.head()
```

```python
data.hist(column='income')
plt.show()
```

```python
data['income'].value_counts()
```

```python
data['income'].value_counts(normalize=True)
```

```python
data.groupby(['workingClass','income']).size().unstack().plot(kind='barh',stacked=True)
plt.rcParams['figure.figsize'] = (18,7)
plt.xticks([])
plt.show()
```

```python
data.groupby(['education','income']).size().unstack().plot(kind='barh',stacked=True)
plt.rcParams['figure.figsize'] = (16,5)
plt.xticks([])
plt.show()
```

```python
data.groupby(['occupation','income']).size().unstack().plot(kind='barh',stacked=True)
plt.rcParams['figure.figsize'] = (18,8)
plt.xticks([])
plt.show()
```

```python
data.groupby(['race','income']).size().unstack().plot(kind='barh',stacked=True)
plt.rcParams['figure.figsize'] = (18,6)
plt.xticks([])
plt.show()
```

```python
data.groupby(['sex','income']).size().unstack().plot(kind='bar',stacked=True)
plt.rcParams['figure.figsize'] = (18,6)
plt.show()
```

```python
data.groupby(['nativeCountry','income']).size().unstack().plot(kind='barh',stacked=True)
plt.rcParams['figure.figsize'] = (18,6)
plt.xticks([])
plt.show()
```

```python
data.groupby(['maritalStatus','income']).size().unstack().plot(kind='barh',stacked=True)
plt.rcParams['figure.figsize'] = (17,3)
plt.xticks([])
plt.show()
```

```python
data.groupby(['relationship','income']).size().unstack().plot(kind='barh',stacked=True)
plt.rcParams['figure.figsize'] = (18,6)
plt.xticks([])
plt.show()
```

```python
data.groupby(['income', pd.cut(data['age'], np.arange(0,100,20))]).size().unstack(0).plot.bar(stacked=True)
plt.rcParams['figure.figsize'] = (18,5)
plt.yticks([])
plt.show()
```

```python
data.groupby(['income', pd.cut(data['hoursPerWeek'], np.arange(0,100,20))]).size().unstack(0).plot.bar(stacked=True)
plt.rcParams['figure.figsize'] = (18,6)
plt.yticks([])
plt.show()
```
